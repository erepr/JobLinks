```python
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import re
import time
import pandas as pd
import csv

PATH = "E:\chromedriver.exe"
driver = webdriver.Chrome(PATH)
```

```python
df = pd.read_csv("finlinks.csv")
df
```

```python
main_data = {'Freelancer Name':[],
             'Job_Links':[]}
```

```python
for i in df['Freelancer_Link']:
    new_link = i + '/reviews'
    driver.get(new_link)
    time.sleep(2)
    
    try:
        name = []
        job_links = []
        try:
            load_more = driver.find_element_by_css_selector("#load-more-feedback").text
            while load_more == 'See More':
                button = driver.find_element_by_xpath('//*[@id="load-more-feedback"]')
                driver.execute_script("arguments[0].click();", button) # Clicks button until this button vanishes
                time.sleep(3)                                            
        except NoSuchElementException as exception:
            flname = driver.find_element_by_css_selector("#freelancer-details > div.p-box.p-identity > div:nth-child(1) > div.profile-avatar > div.profile-avatar__info > h1 > strong").text
            main_data['Freelancer Name'].append(flname) # freelancer name                                          
            links = driver.find_elements_by_css_selector("div > div.avatarinfo > p.copy.u-copySmall.rhythmMargin0 > a")  
            for i in links:
                job_links.append(i.get_attribute("href")) # temp push all joblinks                                                                                          
            
            rem_dup = list(dict.fromkeys(job_links)) # remove duplicates from joblinks
            main_data['Job_Links'].append(rem_dup) # push joblinks into maindata
    except NoSuchElementException as exception:
        pass # skips dead pages
```

```python
df3=pd.DataFrame(main_data)
df3
```
